```python
import pandas as pd
import requests
import io

url = "https://data.ny.gov/resource/vxuj-8kew.csv"

response = requests.get(url, verify=False)
ridership = pd.read_csv(io.StringIO(response.text))

ridership.head()
```

```python
ridership["date"] = pd.to_datetime(ridership["date"])

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(ridership["date"], ridership["subways_total_estimated_ridership"])

plt.title("NYC Subway Ridership Over Time")
plt.xlabel("Date")
plt.ylabel("Total Riders")
plt.xticks(rotation=45)

plt.show()
```

```python
# Smooth the data (7-day rolling average)
ridership["rolling_avg"] = ridership["subways_total_estimated_ridership"].rolling(7).mean()

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

# Plot raw data (light)
plt.plot(ridership["date"], ridership["subways_total_estimated_ridership"], alpha=0.3, label="Daily Ridership")

# Plot smoothed trend (main line)
plt.plot(ridership["date"], ridership["rolling_avg"], linewidth=2, label="7-Day Average")

# Add COVID vertical line
plt.axvline(pd.to_datetime("2020-03-15"), linestyle="--")

# Labels + title
plt.title("NYC Subway Ridership Trends (Pre and Post COVID)", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Total Riders")
plt.xticks(rotation=45)

# Legend
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

# Clean white background
plt.style.use("seaborn-v0_8-whitegrid")

# Plot lines
plt.plot(ridership["date"], ridership["subways_total_estimated_ridership"], alpha=0.2)
plt.plot(ridership["date"], ridership["rolling_avg"], linewidth=3)

# COVID line
covid_date = pd.to_datetime("2020-03-15")
plt.axvline(covid_date, linestyle="--")

# Annotation (THIS is the pro move)
plt.text(covid_date, 4500000, "COVID-19 Begins", rotation=90, va='top')

# Titles and labels
plt.title("NYC Subway Ridership Collapse and Recovery (COVID Impact)", fontsize=16, weight='bold')
plt.xlabel("Date")
plt.ylabel("Total Daily Riders")

plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("ridership_chart.png")  # 👈 ADD THIS LINE
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to your CSV
file_path = "/Users/daneanderson/Downloads/MTA_Daily_Ridership.csv"

# Read the CSV as single column
df = pd.read_csv(file_path, encoding="utf-8-sig")

# Manually split the single column by two or more spaces (tab-like spacing)
df_expanded = df['MTA_Daily_Ridership'].str.split(r'\s{2,}', expand=True)

# Set first row as header
df_expanded.columns = df_expanded.iloc[0]

# Drop the header row
df_expanded = df_expanded[1:]

# Keep only the columns we care about
subway = df_expanded[['Date', 
                      'Subways: Total Estimated Ridership', 
                      'Subways: % of Comparable Pre-Pandemic Day']].copy()

# Rename for convenience
subway.columns = ['Date', 'Ridership', 'Pct_PrePandemic']

# Convert types
subway['Date'] = pd.to_datetime(subway['Date'])
subway['Ridership'] = pd.to_numeric(subway['Ridership'].str.replace(',',''), errors='coerce')
subway['Pct_PrePandemic'] = pd.to_numeric(
    subway['Pct_PrePandemic'].str.replace('%','', regex=False),
    errors='coerce'
)

# Add 30-day rolling average
subway['Rolling_30'] = subway['Pct_PrePandemic'].rolling(30).mean()

# Plot
plt.figure(figsize=(12,6))
plt.plot(subway['Date'], subway['Pct_PrePandemic'], label='Daily % of Pre-Pandemic', alpha=0.5, color='teal')
plt.plot(subway['Date'], subway['Rolling_30'], label='30-Day Rolling Average', color='orange')
plt.title("NYC Subway Recovery as % of Pre-Pandemic Ridership")
plt.xlabel("Date")
plt.ylabel("Percent of Pre-Pandemic Ridership")
plt.legend()
plt.tight_layout()

# Save chart
plt.savefig("/Users/daneanderson/Downloads/subway_recovery_percent_rolling.png", dpi=300, bbox_inches='tight')
plt.show()
```
